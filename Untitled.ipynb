```python
import requests;
import pandas as pd;
```

```python
website_url = requests.get("https://nl.wikipedia.org/wiki/Lijst_van_Belgische_bieren").text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,"lxml")
print(soup.prettify())
```

```python
My_table = soup.findAll("table",{'class':'wikitable sortable'})
```

```python
table = My_table
```

```python
table
```

```python
bieren = []

for rows in table:
    row = rows.findAll("tr");
    for cells in row:
        cell = cells.findAll("td");
        if(len(cell) == 5):
            bier = cell[0].find(text=True)
            soort = cell[1].find(text=True)
            alcohol = cell[2].find(text=True)
            brouwerij = cell[3].find(text=True)
            bieren.append((bier, soort, alcohol, brouwerij ))
```

```python
print(bieren)
```

```python
bierenPanda = pd.DataFrame(bieren)
```

```python
bierenPanda.columns = ["Bier", "Soort", "Alchohol", "Brouwerij"]
```

```python
bierenpd = bierenPanda
```

```python
bierenpd.Brouwerij = bierenpd.Brouwerij.str.strip()
```

```python
alleBrouwerijen = bierenpd.groupby("Brouwerij").Brouwerij.count().reset_index(name="counts")
```

```python
alleBrouwerijen
```

```python
indexBrouwerijen = alleBrouwerijen.index.values
```

```python
teller = 0

for index in indexBrouwerijen:
    indexBrouwerijen[teller] = index + 5000;
    teller = teller + 1;
```
